```python
import time

def linear_search(L, e):
    for i in range(len(L)):
        if (L[i] == e):
            return i

'''
return the runtime of f(arg1, arg2)
- will not be consistent if short time 
  so run multiple times and average it 
  to make it more consistent
'''
def timeit(f, arg1, arg2):
    n_runs = 1000
    start_t = time.time() # amt of seconds since 01/01/1980
    for i in range(n_runs):
        f(arg1, arg2)
    end_t = time.time()
    return (end_t - start_t)/n_runs

timeit(linear_search, [0]*10000, 1) # maximum time of linear search (element does not exist)

lengths = [0, 1, 10, 100, 1000, 10000, 50000]
runtimes = []
for length in lengths:
    runtimes.append(timeit(linear_search, [0]*length, 1))
print(runtimes)

plt.plot(lengths, runtimes)
# proportional, so linear as expected
plt.show()
```

## Complexity and Nested Loops

```python
def f(m, n):
    for i in range(n):
        # repeat n times
        for j in range(m):
            # repeat m times
            pass
def g(n):
    for i in range(n):
        for j in range(n//2):
            pass
```

complexity of f?
- runtime is proportional to number of iterations
- number of iterations is n*m
- O(m*n)

complexity of g?
- runtime is proportional to number of iterations
- number of iterations is n*(n/2) = n^2/2
- O(n^2)
