## The Euler Method

**The Explicit Euler Formula**
<br>



```python
import numpy as np
import matplotlib.pyplot as plt

f0 = -1
dt = 0.1
S = [f0]
step = 0.1
def F(t, S):
    return np.e**(-t)

x = np.arange(0, 1, 0.1)
print(len(x))
for i in range(1, len(x)):
    St = S[i-1] + step*F(x[i-1], S[i-1])
    S.append(St)

# h (or step) = 0.01
step = 0.01
x2 = np.arange(0, 1, 0.01)
S2 = np.zeros(len(x2))
S2[0] = f0
for i in range(1, len(x2)):
    S2[i] = S2[i-1] + step*F(x2[i-1], S2[i-1])

# plot
def f(t):
    return -np.e**(-t)

fig, ax = plt.subplots()
y = f(x)
ax.plot(x, y, label="function")
ax.plot(x, S, '.--', label="h=0.1")
ax.plot(x2, S2, '--', label="h=0.01")
ax.legend()
```
