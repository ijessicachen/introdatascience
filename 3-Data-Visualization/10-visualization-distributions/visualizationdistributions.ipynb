# Bird distributions

```python
import pandas as pd
import matplotlib.pyplot as plt
birds = pd.read_csv('../../data/birds.csv')
birds.head()
```

Show the general way data is distributed via scatterplot.

```python
birds.plot(kind='scatter',x='MaxLength',y='Order',figsize=(12,8))

plt.title('Max Length per Order')
plt.ylabel('Order')
plt.xlabel('Max Length')

plt.show()
```

but histograms are better

## Histograms

You know what a histogram is, courtesy of stats, here is a demonstration with matplotlib:

```python
birds['MaxBodyMass'].plot(kind = 'hist', bins = 10, figsize = (12,12))
plt.show()
```

review: more bins will yield more insight.

```python
birds['MaxBodyMass'].plot(kind = 'hist', bins = 30, figsize = (12,12))
plt.show()
```

For less of a skew, you can change the range. Let's focus on the ton of birds in the low area.

```python
filteredBirds = birds[(birds['MaxBodyMass'] > 1) & (birds['MaxBodyMass'] < 60)]      
filteredBirds['MaxBodyMass'].plot(kind = 'hist',bins = 40,figsize = (12,12))
plt.show()     
```

#### Some experimentation:

```python
birds['MaxLength'].plot(kind = 'hist', bins = 100, figsize = (12,12))
plt.show()
```

Again it's right skewed, this probably means that generally birds are, you know, smaller. <br>
Anyways, let's focus on that data-filled section between maybe 15 and 65.

```python
filteredBirds = birds[(birds['MaxBodyMass'] > 15) & (birds['MaxBodyMass'] < 65)]      
filteredBirds['MaxBodyMass'].plot(kind = 'hist',bins = 50,figsize = (12,12))
plt.show()    
```

Hm, pretty darn uniform. <br>
Okay let's take these two charts from above, and play with some other features:<br>

First: Mark the mean of the graphs with vertical lines.

```python
fig, axs = plt.subplots(2, 1, sharex=True, figsize=(8, 8))
y1 = birds['MaxLength']
y2 = birds[(birds['MaxBodyMass'] > 15) & (birds['MaxBodyMass'] < 65)]

axs[0].hist(y1, bins=100)
axs[1].hist(y2, bins = 50)
```
