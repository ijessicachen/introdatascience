### Task 1: Compute mean values and variance for all values

```python
print("AGE")
mean = df['AGE'].mean()
var = df['AGE'].var()
std = df['AGE'].std()
print(f"Mean = {mean}\nVariance = {var}\nStandard Deviation = {std}\n")

print("SEX")
mean = df['SEX'].mean()
var = df['SEX'].var()
std = df['SEX'].std()
print(f"Mean = {mean}\nVariance = {var}\nStandard Deviation = {std}\n")

print("BMI (body mass index)")
mean = df['BMI'].mean()
var = df['BMI'].var()
std = df['BMI'].std()
print(f"Mean = {mean}\nVariance = {var}\nStandard Deviation = {std}\n")

print("BP (average blood pressure)")
mean = df['BP'].mean()
var = df['BP'].var()
std = df['BP'].std()
print(f"Mean = {mean}\nVariance = {var}\nStandard Deviation = {std}\n")

print("Blood Measurements")
print("Mean = ")
print(df[['S1','S2','S3', 'S4', 'S5', 'S6']].mean())
print("Variance = ")
print(df[['S1','S2','S3', 'S4', 'S5', 'S6']].var())
print("Standard Deviation = ")
print(df[['S1','S2','S3', 'S4', 'S5', 'S6']].std())
print("\n")

print("Y (qualitative measurement of disease progression over 1 year)")
mean = df['Y'].mean()
var = df['Y'].var()
std = df['Y'].std()
print(f"Mean = {mean}\nVariance = {var}\nStandard Deviation = {std}\n")
```

### Task 5: Test the hypothesis that the degree of diabetes progression is different between men and women

Reminder that Y is the best variable to represent degree of diabetes progression.

First we'll test with confidence intervals.

```python
import scipy.stats

# this function taken from the notebook for 1.4 (the lesson for this assignment) 
# in this course
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, h

for p in [0.85,0.9,0.95]:
    m1, h1 = mean_confidence_interval(df.loc[df['SEX']==1,['Y']],p)
    m2, h2 = mean_confidence_interval(df.loc[df['SEX']==2,['Y']],p)
    print(f'Conf={p:.2f}, Sex 1 Y: {m1-h1[0]:.2f}..{m1+h1[0]:.2f}, Sex 2 Y: {m2-h2[0]:.2f}..{m2+h2[0]:.2f}')
```

So there is some overlap, and as the confidence interval increases it eventaully turns out that most of it is overlapping in confidence interval 0.95. I don't think this supports the hypothesis that the degree of progression is different, but to really make sure, let's use the student t-test method.

```python
from scipy.stats import ttest_ind

tval, pval = ttest_ind(df.loc[df['SEX']==1,['Y']], df.loc[df['SEX']==2,['Y']],equal_var=False)
print(f"T-value = {tval[0]:.2f}\nP-value: {pval[0]}")
```

So there is a lot of overlap even if the p-value is fairly small. I will also say that in many cases, the p-value should be below 0.05 for results truly to be statistically significant, so I will conclude that this hypothesis if __false__.
